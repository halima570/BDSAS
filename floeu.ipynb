```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
credit=ctrl.Antecedent(np.arange(2000,50000,1),'credit')
```

```python
credit['petit']=fuzz.trapmf(credit.universe,[2000,2000,15000,21000])
credit['moyen']=fuzz.trapmf(credit.universe,[15000,21000,31000,37000])
credit['grand']=fuzz.trapmf(credit.universe,[31000,37000,50000,50000])
credit.view()
```

```python
montant_credit=ctrl.Antecedent(np.arange(0,100000,1),'montant_credit')
montant_credit['trespetit'] = fuzz.trimf(montant_credit.universe, [0, 0, 20000])
montant_credit['petit'] = fuzz.trimf(montant_credit.universe, [0, 20000, 40000])
montant_credit['moyen'] = fuzz.trimf(montant_credit.universe, [20000, 40000, 60000])
montant_credit['grand'] = fuzz.trimf(montant_credit.universe, [40000, 60000, 80000])
montant_credit['tresgrand'] = fuzz.trimf(montant_credit.universe, [60000, 80000, 100000])
montant_credit.view()
```

```python
montant_credit['trespetit'] = fuzz.trapmf(montant_credit.universe, [0, 0, 15000,20000])
montant_credit['petit'] = fuzz.trapmf(montant_credit.universe, [15000, 20000, 35000,45000])
montant_credit['moyen'] = fuzz.trapmf(montant_credit.universe, [35000, 45000, 55000,65000])
montant_credit['grand'] = fuzz.trapmf(montant_credit.universe, [55000, 65000, 75000,85000])
montant_credit['tresgrand'] = fuzz.trapmf(montant_credit.universe, [75000, 85000, 100000,100000])
montant_credit.view()
```
