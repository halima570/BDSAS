```python
# read in all our data
#df =pd.read_excel("C:/Users/Administrateur/python/rachid/auto evaluation/questionnaire.xlsx")
#print(df.shape)
df = pd.read_csv("E:/gouvernance de donneees/2021/support de cours/dataset_titanic/train.csv")
print(df.shape)
```

```python
df['Cabin']=df['Cabin'].replace(np.NaN,'cab0')
print(df['Cabin'][:20])
```

```python
df.loc[(df.Cabin != 'cab0'), 'Cabin']='cab1'
print(df['Cabin'][:20])
```

```python
cab = pd.get_dummies(df['Cabin'],drop_first=False)
df.drop(['Cabin'],axis=1,inplace=True)
df = pd.concat([df,cab],axis=1)
df.head(4)
```

```python
emb = pd.get_dummies(df['Embarked'],drop_first=False)
df.drop(['Embarked'],axis=1,inplace=True)
df = pd.concat([df,emb],axis=1)
df.head(4)
```

```python
df["Age"] = df["Age"].interpolate(method='linear', limit_direction='forward', axis=0)
#print(df['Age'][:20])
```

```python
sex = pd.get_dummies(df['Sex'],drop_first=False)
sex
```

```python
df.drop(['Sex'],axis=1,inplace=True)
df = pd.concat([df,sex],axis=1)
df
```
