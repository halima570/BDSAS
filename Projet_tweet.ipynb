# Scraping avec ntscraper

```python
from ntscraper import Nitter
scraper = Nitter()
```

```python
import pandas as pd
def get_tweets(terme,mode,nombre):
    tweets = scraper.get_tweets(terme,mode=mode,number=nombre)
    final_tweets = []

    for tweet in tweets['tweets']:
        data_tweet = [tweet['link'], tweet['text'], tweet['date'], tweet['stats']['likes'], tweet['stats']['comments']]
        final_tweets.append(data_tweet)

    data_df = pd.DataFrame(final_tweets, columns=['link', 'text', 'date', 'Likes', 'Comments'])
    return data_df
```

```python
tweet = get_tweets("ChatGPT",'hashtag',100)
```

# Classification

```python
from textblob import TextBlob
import pandas as pd

df = pd.read_csv('ChatGPT1.csv')

# Function to get sentiment label
def get_sentiment_label(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    
    # Determine sentiment label based on polarity score
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to each text in the DataFrame
df['sentiment'] = df['text'].apply(get_sentiment_label) 
```
